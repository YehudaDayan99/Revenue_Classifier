```python
import os
from pprint import pprint

# Required: SEC compliance user agent (must include your email)
os.environ.setdefault("SEC_USER_AGENT", "RevenueSegBot/0.1 (josh.dayan@gmail.com)")

# Required for ReAct pipeline
# IMPORTANT: do NOT hardcode API keys into notebooks/repos.
# Set OPENAI_API_KEY in your environment before running.
if not os.getenv("OPENAI_API_KEY"):
    raise RuntimeError(
        "OPENAI_API_KEY is not set. In PowerShell run: $env:OPENAI_API_KEY='...'; then restart the kernel."
    )

from revseg.pipeline import run_pipeline

tickers = ["AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "META", "AVGO", "TSLA", "JPM", "XOM"]

report = run_pipeline(
    tickers=tickers,
    model="gpt-4.1-mini",
    max_react_iters=3,
    out_dir="data/outputs",
)

# Compact summary
summary = {
    t: {
        "ok": r.get("ok"),
        "segment_year": r.get("segment_year"),
        "n_segments": r.get("n_segments"),
        "errors": r.get("errors"),
    }
    for t, r in report.get("tickers", {}).items()
}

pprint(summary)
print("\nWrote outputs under:", report.get("outputs_dir"))
```
